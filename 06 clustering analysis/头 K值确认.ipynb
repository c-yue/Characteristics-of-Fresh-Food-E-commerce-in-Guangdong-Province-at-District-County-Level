# 1  手肘法

```python
import pandas as pd
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt
 
df_features = pd.read_excel(r'F:/Homework/数据挖掘/作业一/聚类/chapter7/demo/data/tmp/数据预处理/zz.xls',encoding='gbk') # 读入数据
'利用SSE选择k'
SSE = []  # 存放每次结果的误差平方和
for k in range(2,9):
    estimator = KMeans(n_clusters=k)  # 构造聚类器
    estimator.fit(df_features[['ZL','ZR','ZF','ZM','ZC']])
    SSE.append(estimator.inertia_)
X = range(2,9)
plt.xlabel('k')
plt.ylabel('SSE')
plt.plot(X,SSE,'o-')
plt.show()
```

# 2  轮廓系数法

```python
import pandas as pd
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score
import matplotlib.pyplot as plt
 
df_features = pd.read_excel(r'F:/Homework/数据挖掘/作业一/聚类/chapter7/demo/data/tmp/数据预处理/zz.xls',encoding='gbk')
Scores = []  # 存放轮廓系数
for k in range(2,9):
    estimator = KMeans(n_clusters=k)  # 构造聚类器
    estimator.fit(df_features[['ZL','ZR','ZF','ZM','ZC']])
    Scores.append(silhouette_score(df_features[['ZL','ZR','ZF','ZM','ZC']],estimator.labels_,metric='euclidean'))
X = range(2,9)
plt.xlabel('k')
plt.ylabel('轮廓系数')
plt.plot(X,Scores,'o-')
plt.show()
```

```python
import pandas as pd
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score
import matplotlib.pyplot as plt
 
df_features = pd.read_csv('F:\\Homework\\数据挖掘\\作业一\\聚类\\new\\data\\iris.csv',encoding='gbk')  # 读入数据
Scores = []  # 存放轮廓系数
for k in range(2,9):
    estimator = KMeans(n_clusters=k)  # 构造聚类器
    estimator.fit(df_features[['Sepal.Length','Sepal.Width','Petal.Length','Petal.Width']])
    Scores.append(silhouette_score(df_features[['Sepal.Length','Sepal.Width','Petal.Length','Petal.Width']],estimator.labels_,metric='euclidean'))
X = range(2,9)
plt.xlabel('k')
plt.ylabel('轮廓系数')
plt.plot(X,Scores,'o-')
plt.show()
```
